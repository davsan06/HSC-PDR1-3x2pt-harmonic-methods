```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
As = np.random.uniform(low=3.162277660168379e-09,high=0.0001,size=1000)
log10As = np.random.uniform(low=1.5,high=6,size=1000)
```

```python
10 ** 1.5 / 10 ** 10
```

```python
10 ** 6 / 10 ** 10
```

```python
np.log10(10**10 * 1.5)
```

```python
fig=plt.hist(As,bins=int(np.sqrt(len(As))),histtype='step')
```

```python
fig = plt.hist(log10As,bins=int(np.sqrt(len(log10As))),histtype='step')
```

```python
As_conv = 10 ** log10As / 10 ** 10
```

```python
fig = plt.hist(As_conv,bins=int(np.sqrt(len(As_conv))),histtype='step')
```
